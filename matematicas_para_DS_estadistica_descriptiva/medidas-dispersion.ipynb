# Medidas de dispersión

continuamos con el dataset https://www.kaggle.com/lepchenkov/usedcarscatalog

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('cars.csv')
```

```python
df['price_usd'].std()  # ? .std() da la desviacion estandar
```

```python
  # ? Rango, valor max - valor min 

rango = df['price_usd'].max() - df['price_usd'].min()  # toma el valor maximo y minimo de 'price_usd' y lo resta, dandonos el rango
rango
```

```python
  # ? Quartiles 

median = df['price_usd'].median()
Q1 = df['price_usd'].quantile(q=0.25)  # ? .quantile(q=0.25), indica que se debe tomar el primer 25% de los valores (el primer cuarto)
Q3 = df['price_usd'].quantile(q=0.75)
min_val = df['price_usd'].quantile(q=0)  # Se indica que tome el valor 0 con quantile(q=0)
max_val = df['price_usd'].quantile(q=1) 

print(min_val, Q1, median, Q3, max_val)
```

```python
iqr = Q3 - Q1
iqr
```

### Límites para detección de outliers (datos simetricamente distribuidos)

Datos entre $$Q_1 -1.5 \times \text{IQR}$$ y $$Q_3 + 1.5 \times \text{IQR}$$

```python
minlimit = Q1 - 1.5 * iqr  # asi vemos el limite minimo (antes de los outliers de abajo)
maxlimit = Q3 + 1.5 * iqr  # asi vemos el limite maximo (antes de los outliers de arriba)

print(minlimit, maxlimit)
```
