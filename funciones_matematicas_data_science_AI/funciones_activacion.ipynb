### Funcion de escalon o Heaviside 
$$H(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

```python
def H(x):
  Y = np.zeros(len(x))
  for idx, x in enumerate(x):
    if x >= 0:
      Y[idx]=1
  return Y

N=1000

y = H(x)

plt.plot(x, y)
plt.grid()
```

### Función sigmoide

$$f(x)=\frac{1}{1-e^{-x}}$$

```python
def f(x):
  return 1/(1 + np.exp(-x))  # ? exp(), eleva el numero que se le pase por parametro por el numero "e" que es aprox: 2.718

N = 1000

y = f(x)

plt.plot(x, y)
plt.grid()
```
