Eliminar datos en pandas 

Eliminando columnas, Drop colummns 

```python
  # ? .drop() elimina la columna, 'Genre': nombre de la columna a eliminar. axis=1, indica que si es a una columna que aputamos
  # ? .drop() NO altera el data frame original
df_books.drop('Genre', axis=1).head(4) 
```

```python
df_books.head(4)
```

```python
df_books.drop('Genre', axis=1, inplace=True)  # todo: inplace=true, signifca que los cambios SI afectara el data frame original
df_books.head(4)
```

```python
df_books = df_books.drop('Year', axis=1)  # ? Esta es otra forma de eliminar columnas
df_books.head(4)
```

```python
del df_books['Price'] # ? Este es un metodo propio de python (no recomendado) pero es otra forma de eliminar columnas
```
