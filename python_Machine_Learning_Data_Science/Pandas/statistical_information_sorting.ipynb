```python
df['total_bill'].between(10, 20, inclusive=True)  # ? Muestra si el valor de 'total_bill' esta entre 10 y 20 
```

```python
df[df['total_bill'].between(10, 20, inclusive=True)]  # Regresa un nuevo dataframe con valores de "total_bill" que solo enten entre 10 y 20
```

```python
df.nlargest(10, 'tip')  # ? Muestra las 10 columnas que tienen los 10 valores de 'tip' mas altos
```

```python
df.sort_values('tip', ascending=False).iloc[0:10]  # Esto hace lo mismo de arriba, pero la version de arriba es mejor
```

```python
df.nsmallest(10, 'tip')  # ? Este muestra los 10 valores mas pequeños de todos
```

```python
df.sample(5)  # ? Devuelve 5 valores random como 'muestra' del dataframe
```

```python
df.sample(frac=0.1)  # ? Devuelve 'muestras' que sean el 10% de las totales del dataframe 
```
