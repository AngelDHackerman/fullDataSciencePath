```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
# analizando la relacion entre el total de las cuentas y las propinas
# hue='day', podemos ver la frecuencia de los dias 
# style='time', si es 'lunch' es un marcador 'O', si el tiempo es 'dinner' sera un marcador 'X'
# size='size', el tamao del marcador sera segun el tamao de la mesa que fue servida e.g: 2 personas, 4 personas.
# bbox_to_anchor=(1.12, 0.5) es lo que nos pone la leyenda al lado derecho de la tabla

sns.scatterplot(data=tips, x= 'total_bill', y='tip', hue='day', style='time', size='size')
plt.legend(loc='center', bbox_to_anchor=(1.12, 0.5))
plt.show()
```

```python
# si el label es 'lunch' su marker sera 'delta', si el label es 'Dinner' su marker sera 'square'
markers= {'Lunch': 'D', 'Dinner': 's'}
sns.scatterplot(data=tips, x= 'total_bill', y='tip', hue='day', style='time', size='size', markers=markers)
plt.legend(loc='center', bbox_to_anchor=(1.12, 0.5))
plt.show()
```
