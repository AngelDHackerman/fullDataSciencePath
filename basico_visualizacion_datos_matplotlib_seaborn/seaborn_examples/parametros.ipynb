```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')  # ? .load_dataset('tips'), es un conjunto de datasets precargados que ya trae seaborn
```

```python
tips
```

```python
sns.displot(data=tips, x='total_bill')  # muestra la frecuencia de cuando se entregan propinas, basado en el total de la cuenta
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', y='tip')  # muestra la frecuencia de cuando se entregan propinas, basado en el total de la cuenta
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', y='tip', hue='sex')  # ? hue='', es una forma de especificar una tercera variable (una variable categorica)
plt.show()
```

```python
# lo mismo de arriba pero solo con la variable en el eje 'x'
sns.displot(data=tips, x='total_bill', hue='sex')  # ? hue='', es una forma de especificar una tercera variable (una variable categorica)
plt.show()
```

```python
  # todo: .displot(la data, eje x, eje y, como sera segmentado)
  # ? kind='', muestra el tipo de grafico queremos trabajar. 
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde', legend=True, palette='dark', alpha=0.35)  
plt.show()
```
