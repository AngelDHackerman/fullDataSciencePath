```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips.head(2)
```

```python
sns.countplot(data=tips, x='day', hue='sex')  # ? countplot(), cuenta cuantos registros tenemos en el eje de X, y tambien segun la variable categorica
plt.show()
```

```python
sns.countplot(data=tips, y='day', hue='sex')
plt.show()
```

```python
  # ? stripplot, hace un grafico categorico tipo scatter, pero reduciendo el overlapping 
  # ? dodge=true, este parametro hace que las burbujas de 'sex' no esten juntas para visualizar mejor la info
plt.figure(figsize=(6, 6))
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)  
plt.show()
```

```python
  # ? swarmplot(), es similar a striplot() pero aqui se crea un eje horizontal donde se concentran mas los datos
plt.figure(figsize=(6, 6))
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)  
plt.show()
```

```python
sns.boxplot(data=tips, y='total_bill') # se ve la media de las propinas, y los outliers
plt.show()
```

```python
  # ? boxplot(), las lineas de enmedio nos permiten ver la media de 'tips', segun el 'sex' por hombres y mujeres en cada diferente dia
  # * los 'diamantes' de arriba son los outliers, que se salen del promedio de las propinas
plt.figure(figsize=(6, 6))
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)  
plt.show()
```

```python
  # ? plt.legend(loc='upper left'), esto mueve la leyenda de male, female... a la izquierda superior
plt.figure(figsize=(8, 8))
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)  
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, palette='dark:0', marker='<')  
plt.legend(loc='upper left')
plt.show()
```

```python
  # ? En el violinplot(), el lado mas ancho de la figura muestra donde esta la concentracion de los datos
  # ? El parametro split=True, une los 2 graficos en uno solo.
plt.figure(figsize=(8, 8))
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', palette='pastel', split=True)  
plt.legend(loc='upper left')
plt.show()
```

```python
sns.violinplot(data=tips, y='total_bill')
```

```python
  # ? catplot(), es el 'comodin' para los graficos, en el parametro kind='' le podemos decir que grafico preferimos 
  # ? el parametro col='time', muestra como se comporta el grafico en el almuerzo como en la cena
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, kind='swarm', col='time')
plt.show()
```
