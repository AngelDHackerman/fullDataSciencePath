```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = np.linspace(0, 5, 11) 
y = np.sin(x)  # ? .sin() es la funcion de seno.  
```

```python
fig, axes = plt.subplots()  # crea una figura (canvas) con axes (lienzo) dentro
axes.plot(x, y, 'b')  # .plot() es un grafico de lineas
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)  # ? .subplots() crea multiples graficos en la misma figura. (1 fila y 2 columnas)
axes[0].plot(x, y, 'b')  # Aqui se pueden ingresar a las figuras como si fueran un array
axes[1].plot(y, x, 'r')
```

#### Otra forma de acceder a los graficos dentro de la figura: 

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)  # (ax1, ax2), se pueden declarar asi para evitar ingresar a ellos como un array. 
ax1.plot(x, y, 'b')
ax2.plot(y, x, 'r')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=4)
axes[0, 0].plot(x, np.cos(x))  # ? axes[filas, columnas], fila 0 / columna 0
axes[0, 1].plot(x, np.sin(x), 'y')
axes[0, 2].plot(x, np.tan(x), 'r')
axes[0, 3].plot(x, np.cos(x) ** 2)
```
