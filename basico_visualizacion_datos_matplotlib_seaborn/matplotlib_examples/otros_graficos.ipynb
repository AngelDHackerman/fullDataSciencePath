```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1, 50, 100)
```

## Histograma, grafico para analizar distribucion y frecuencia:

```python
  # ? bins, es la propiedad principal de los histogramas, aqui se puede ver la frecuencia de los datos. 
plt.hist(data, bins=15)  
plt.show()
```

```python
  # ? histtype='', le podemos decir la forma del histograma
plt.hist(data, bins=15, histtype='step')  
plt.show()
```

## Boxplot, grafico de cajas

```python
# boxplot, muestra donde se encuentra el percentil 50 de nuestros datos
# vert=False, hace la grafica de manera horizontal
# patch_artist, cambia el color de background para hacer mas entendible la grafica
# notch=True, concentra la visualizacion de los datos 
# showfliers=False, Si tenemos un dato FUERA DE LO COMUN, esta opcion no la muestra en la grafica 

data = np.append(data, 150)
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=False)
plt.show()
```

## Scatter, grafico de dispersion: 

```python
N = 50 
x = np.random.rand(N)  # creamos un vector (array) con 50 valores random
y = np.random.rand(N)
area = (30 * np.random.rand(N)) ** 2
colors = np.random.rand(N)
```

```python
  # ? s='', sirve para especificar el tamao de los puntos en la grafica.
  # ? maker='', podemos cambiar la forma del marcador, por defecto es 'o'
  # ? alpha=0.5, sirve para cuando los graficos estan superpuestos, esto nos da cierta transparencia
plt.scatter(x, y, s=area, c=colors, marker='o', alpha=0.5)
plt.show()
```
