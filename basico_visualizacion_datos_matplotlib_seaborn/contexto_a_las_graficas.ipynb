```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11) # crea un array de 10 elementos
y = np.sin(x)  # crea otro array de 10 elementos pero con su producto de seno.
```

```python
fig, axes = plt.subplots(1, 2, figsize=(7, 5)) # 1 fila, 2 columnas # ? figsize=(widht, height)
axes[0].plot(x, y, label="$sin(x)$")  # ? Podemos agrear el label dentro de la figuara.    el uso de $ $ es para crear notaciones matematicas
axes[0].set_title('Relacion X - Y') # ? .set_title(), esta propiedad nos permite agregarle un 'label' o un titulo al grafico
axes[0].set_xlabel('X') # ? Agregamos un 'label' al eje de X y se puede hacer lo mismo con el eje Y1
axes[0].set_ylabel('Y')
axes[0].legend()  # ? .legend() nos muestra las labels dentro de la grafica

axes[1].plot(y, x, 'r')
axes[1].set_title('Relacion Y - X')
axes[1].set_ylabel('X')
axes[1].set_xlabel('Y')
```

#### Esta es la forma de hacerlo, sin usar el metodo orientado a objetos: 

```python
plt.plot(x, y, label='sin(x)')
plt.title('Este es un titulo')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='lower left')  # ? loc='lower right'; Aqui le indicamos la ubicacion de la label
```
