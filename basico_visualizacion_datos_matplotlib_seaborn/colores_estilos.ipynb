```python
import matplotlib.pyplot as ptl 
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
```

```python
print(ptl.style.available)  # ? Imprime los estilos diponibles por defecto 
```

```python
ptl.style.use('bmh')  # ? Asi podemos usar cualquier estilo disponible de los que se mostraron antes 
```

## Modificacion de tipo Matlab:

```python
fig, ax = ptl.subplots(figsize=(8, 8))

ax.plot(x, x+1, 'r--')
ax.plot(x, x+2, 'b-')
ax.plot(x, x+3, 'y.-')
ax.plot(x, x+4, 'go:')
```

## Modificaciones con Pyplot: 

```python
fig, ax = ptl.subplots(figsize=(8, 8))

ax.plot(x, x+1, color = 'red', alpha=0.6, linewidth=5, linestyle='-', marker='o')  # alpha, da transparencia. 
ax.plot(x, x+2, color = '#FFFFFF', linewidth=10, linestyle='--', marker='x')  # linewidth, es el ancho de la linea
ax.plot(x, x+3, color = '#66FF66', linewidth=3, linestyle='dashed', marker='v', markersize=12, markerfacecolor='blue')  # markersize/markerfacecolor: cambian el tamaño y color del marcador
ax.plot(x, x+4, color = 'black', linewidth=2, linestyle=':', marker='p')
```
