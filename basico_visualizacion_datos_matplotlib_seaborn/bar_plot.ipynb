```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
country = ['INDIA', 'JAPAN', 'MEXICO', 'COLOMBIA', 'GERMANY']
population = [1000, 800, 900, 1000, 300]
```

```python
# .bar(), crea un grafico de barras .bar(eje x, eje y, ancho de barras)
# La función xticks() de matplotlib se utiliza para establecer o cambiar las etiquetas de las marcas de tiempo en el eje x de un gráfico de Matplotlib. 
# Por ejemplo, en un gráfico de barras, xticks() puede usarse para cambiar el texto de las etiquetas de los valores en el eje x. 
# La función xticks() acepta una lista de valores y una lista de etiquetas para esos valores.

plt.bar(country, population, width=0.6, color=['red', 'blue']) 
  # ? np.arange(5): lo creamos para tener el mismo numero de elementos que el eje x original. 
plt.xticks(np.arange(5), ('India', 'Japon', 'Mexico', 'Colombia', 'Alemania'), rotation=45) # rotation, cambia el angulo de las etiquetas del eje X 
plt.show()
```

```python
plt.barh(country, population)  # ? Es la misma grafica de arriba pero de manera horizontal
plt.show()
```
